```python
%matplotlib inline
import numpy as np
import cv2
from matplotlib import pyplot as plt
from skimage.morphology import extrema
from skimage.morphology import watershed as skwater

def ShowImage(title,img,ctype):
  plt.figure(figsize=(10, 10))
  if ctype=='bgr':
    b,g,r = cv2.split(img)       # get b,g,r
    rgb_img = cv2.merge([r,g,b])     # switch it to rgb
    plt.imshow(rgb_img)
  elif ctype=='hsv':
    rgb = cv2.cvtColor(img,cv2.COLOR_HSV2RGB)
    plt.imshow(rgb)
  elif ctype=='gray':
    plt.imshow(img,cmap='gray')
  elif ctype=='rgb':
    plt.imshow(img)
  else:
    raise Exception("Unknown colour type")
  plt.axis('off')
  plt.title(title)
  plt.show()
```

```python
img           = cv2.imread("/content/gdrive/My Drive/FYP/data/test/0/benign (370).png")
gray          = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
img_median = cv2.medianBlur(gray, 5)
ShowImage('Median filter', img_median, 'gray')
```

```python
equalized = cv2.equalizeHist(img_median)
ShowImage('Histogram Equa',equalized, 'gray')
```
